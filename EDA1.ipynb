## Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/ravil/Jupyter note book/assigment/eda 1/Cardiotocographic.csv")
```

```python
data.isnull().sum()
```

```python
# Impute missing values with the mean
df = data.fillna(data.mean())

# Check if there are any more missing values
df.isnull().sum()
```

```python
# Statistical summary
df.describe().T
```

```python
import matplotlib.pyplot as plt

# Plot histograms for all numerical columns
df.hist(bins=20, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Create boxplots for each numerical variable
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar plot for the 'NSP' column
sns.countplot(x='NSP', data=df)
plt.title('Distribution of NSP')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
